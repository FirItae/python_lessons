
## Задача 1: Переводчик пиратского языка
Создай словарь для перевода обычных слов на пиратский язык. 
Например:
- "привет" → "йо-хо-хо"
- "друг" → "морской волк"
(Придумай еще 5 слов)
Напиши программу-переводчик, которая превратит обычный текст в пиратскую речь!







## Задача 2: Инвентарь героя
Создай словарь для хранения инвентаря героя в компьютерной игре.
Каждый предмет должен иметь:
- название
- тип (оружие/броня/зелье)
- редкость (обычный/редкий/эпический)
- характеристики (урон для оружия, защита для брони, эффект для зелья)

Напиши программу, которая позволит:
1. Добавлять новые предметы в инвентарь
2. Удалять предметы из инвентаря
3. Показывать полный список предметов с их характеристиками
4. Искать предметы по типу или редкости
